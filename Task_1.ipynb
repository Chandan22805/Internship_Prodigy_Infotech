```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Load Titanic dataset

```python
df_titanic = sns.load_dataset('titanic')
```

# Bar chart: Frequency of passengers by class

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='class', data=df_titanic)
plt.title('Frequency Distribution of Passengers by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

# Histogram: Distribution of passenger ages


```python
plt.figure(figsize=(10, 6))
sns.histplot(df_titanic['age'].dropna(), bins=10, kde=True)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
